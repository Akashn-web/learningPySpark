## Bonus Question!
#### There are too many decimal places for mean and stddev in the describe() dataframe. Format the numbers to just show up to two decimal places. Pay careful attention to the datatypes that .describe() returns, we didn't cover how to do this exact formatting, but we covered something very similar. [Check this link for a hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.Column.cast)

If you get stuck on this, don't worry, just view the solutions.

```python
from pyspark.sql.functions import format_number, mean, stddev, countDistinct, year, month, count
```

```python
cols = df.columns
for i in cols[1:]:
  mean_val = df.select(mean(i).alias('mean'+str(i)))
  std_val = df.select(stddev(i).alias('std'+str(i)))
  mean_val.select(format_number('mean'+str(i), 2)).show()
  std_val.select(format_number('std'+str(i), 2)).show()
  
# mean_val = df.select(mean('Adj Close').alias('mean1'))
# std_val = df.select(stddev('Adj Close').alias('std'))

# mean_val.select(format_number('mean1', 2)).show()
# std_val.select(format_number('std', 2)).show()

# ['Open', 'High', 'Low', 'Close', 'Volume', 'Adj Close']
```

```python
res = df.describe()

res.select(res['summary'], 
          format_number(res['Open'].cast('float'),2).alias('Open'),
          format_number(res['Close'].cast('float'),2).alias('Close'),
          format_number(res['High'].cast('float'),2).alias('High'),
          format_number(res['Low'].cast('float'),2).alias('Low'),
          res['Volume'].cast('int').alias('Volume')
          ).show()
```

#### Create a new dataframe with a column called HV Ratio that is the ratio of the High Price versus volume of stock traded for a day.

```python
new_df = df.withColumn('HV Ratio', df['High']/df['Volume'])
new_df.select('HV Ratio').show(5)
```

#### What day had the Peak High in Price?

```python
max_high = df.agg({'High':'max'}).collect()
max_high_row = df.filter(df['High'] == max_high[0][0])
max_high_row.select('Date').show()
```

#### What is the mean of the Close column?

```python
close_mean = df.agg({'Close':'mean'}).collect()
print("Mean Value of Close Price",close_mean[0][0])
```

#### What is the max and min of the Volume column?

```python
max_vol = df.agg({'Volume':'max'}).collect()
min_vol = df.agg({'Volume':'min'}).collect()

print("Max Val of Volume:", max_vol[0][0])
print("Min Val of Volume:", min_vol[0][0])
```

#### How many days was the Close lower than 60 dollars?

```python
close_limit = df.filter(df['Close'] < 60)
# print((close_limit[0]))
close_limit.select(count('Close')).show()
```

#### What percentage of the time was the High greater than 80 dollars ?
#### In other words, (Number of Days High>80)/(Total Days in the dataset)

```python
gt_80 = df.filter(df['High'] > 80).collect()
print("Percentage of the time was the High greater than 80 dollars:",len(gt_80)*100/df.count(),"%")
```

#### What is the Pearson correlation between High and Volume?
#### [Hint](http://spark.apache.org/docs/latest/api/python/pyspark.sql.html#pyspark.sql.DataFrameStatFunctions.corr)

```python
from pyspark.sql.functions import corr, avg
```

```python
df.select(corr("High", "Volume")).show()
```

#### What is the max High per year?

```python
more_df = df.withColumn('Year', year(df['Date']))
more_df.groupBy('Year').max().select(['Year','max(High)']).show()
```

#### What is the average Close for each Calendar Month?
#### In other words, across all the years, what is the average Close price for Jan,Feb, Mar, etc... Your result will have a value for each of these months.

```python
one_more_df = df.withColumn('Month', month(df['Date']))
one_more_df.groupBy('Month').avg().select(['Month', 'avg(Close)']).orderBy('Month').show()
```
