```
from pyspark.ml.regression import LinearRegression
```

```
training = spark.read.format('libsvm').load('/FileStore/tables/sample_linear_regression_data.txt')
```

```
training.show(10)
```

```
lr = LinearRegression(featuresCol='features', labelCol='label', predictionCol='prediction')
```

```
lrModel = lr.fit(training)
```

```
lrModel.coefficients # weight or coefficients of the features.
```

```
lrModel.intercept # intercept of the best fit line
```

```
training_summary = lrModel.summary
```

```
print(training_summary.r2)
print(training_summary.rootMeanSquaredError)
```

```
all_data = spark.read.format('libsvm').load('/FileStore/tables/sample_linear_regression_data.txt')
```

```
train_data, test_data = all_data.randomSplit([.7,.3])
```

```
train_data.describe().show()
```

```
correct_model = lr.fit(train_data)
```

```
test_results = correct_model.evaluate(test_data)
```

```
print(test_results.rootMeanSquaredError)
print(test_results.r2adj)
```

```
unlabled_data = test_data.select('features')
```

```
predictions = correct_model.transform(unlabled_data)
```

```
predictions.show()
```
